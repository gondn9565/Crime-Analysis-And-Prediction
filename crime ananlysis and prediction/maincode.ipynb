```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.covariance import GraphicalLasso
import warnings
warnings.filterwarnings('ignore')


try:
    from IPython.display import display
except ImportError:
    # Fallback for regular Python environment
    def display(df):
        print(df)

# ===== CRIME ANALYSIS CLASS =====
class CrimeAnalysis:
    def __init__(self, file_path):
        print(f"Loading data from: {file_path}")
        try:
            self.df = pd.read_csv(file_path, encoding='utf-8')
            self.preprocessed = False
            print(f"Data loaded successfully. Shape: {self.df.shape}")
            print(f"Columns: {list(self.df.columns)}")
        except FileNotFoundError:
            print(f"ERROR: File '{file_path}' not found!")
            print("Please check:")
            print("   - File exists in the current directory")
            print("   - File name is correct (case-sensitive)")
            print("   - You're running the script from the right folder")
            # Create empty dataframe to prevent further errors
            self.df = pd.DataFrame()
        except Exception as e:
            print(f"ERROR loading file: {e}")
            self.df = pd.DataFrame()
        
    def preprocess_data(self):
        """Preprocess the crime dataset with correct date formats"""
        # Check if data was loaded successfully
        if self.df.empty:
            print("No data to preprocess. Please check file path.")
            return
        
        print("Preprocessing data...")
        
        # Debug: Show sample of date columns to understand format
        print("\nSample date values:")
        if 'Time of Occurrence' in self.df.columns:
            print("Time of Occurrence sample:", self.df['Time of Occurrence'].head(3).tolist())
        if 'Date of Occurrence' in self.df.columns:
            print("Date of Occurrence sample:", self.df['Date of Occurrence'].head(3).tolist())
        
        # Convert date columns to datetime with DAY FIRST (dd-mm-yyyy)
        date_columns = ['Date Reported', 'Date of Occurrence', 'Date Case Closed']
        for col in date_columns:
            if col in self.df.columns:
                print(f"Converting {col}...")
                # Use dayfirst=True for dd-mm-yyyy format
                self.df[col] = pd.to_datetime(self.df[col], dayfirst=True, errors='coerce')
                # Check for conversion errors
                null_count = self.df[col].isnull().sum()
                if null_count > 0:
                    print(f"Warning: {null_count} null values in {col} after conversion")
        
        # Extract time-based features - FIXED VERSION
        print("Processing Time of Occurrence...")
        try:
            # Try multiple possible formats for time column
            self.df['Hour'] = pd.to_datetime(self.df['Time of Occurrence'], dayfirst=True, errors='coerce').dt.hour
            # Fill any missing hours with a default value
            self.df['Hour'].fillna(12, inplace=True)
            print(f"Hour extraction completed. Missing values: {self.df['Hour'].isnull().sum()}")
        except Exception as e:
            print(f"Error extracting hour: {e}")
            # Alternative approach: extract hour from string if datetime conversion fails
            try:
                self.df['Hour'] = self.df['Time of Occurrence'].str.split(':').str[0].str[-2:].astype(int)
                print("Hour extracted from string")
            except:
                self.df['Hour'] = 12  # Default value
                print("Using default hour value")
        
        # Extract date features
        if 'Date of Occurrence' in self.df.columns:
            self.df['DayOfWeek'] = self.df['Date of Occurrence'].dt.dayofweek
            self.df['Month'] = self.df['Date of Occurrence'].dt.month
            self.df['DayOfMonth'] = self.df['Date of Occurrence'].dt.day
            print("Date features extracted")
        
        # Calculate reporting delay
        if all(col in self.df.columns for col in ['Date Reported', 'Date of Occurrence']):
            self.df['Reporting_Delay_Days'] = (self.df['Date Reported'] - self.df['Date of Occurrence']).dt.days
            print("Reporting delay calculated")
        
        # Handle missing values
        print("Handling missing values...")
        if 'Weapon Used' in self.df.columns:
            self.df['Weapon Used'].fillna('Unknown', inplace=True)
        if 'Victim Age' in self.df.columns:
            self.df['Victim Age'].fillna(self.df['Victim Age'].median(), inplace=True)
        if 'Victim Gender' in self.df.columns:
            self.df['Victim Gender'].fillna('Unknown', inplace=True)
        
        # Create binary target for crime prediction
        if 'Crime Domain' in self.df.columns:
            self.df['Violent_Crime'] = self.df['Crime Domain'].apply(
                lambda x: 1 if 'Violent' in str(x) else 0
            )
            violent_count = self.df['Violent_Crime'].sum()
            print(f"Violent crime target created: {violent_count} violent crimes")
        
        # Encode categorical variables
        categorical_cols = ['City', 'Crime Description', 'Victim Gender', 'Weapon Used', 'Crime Domain']
        self.encoders = {}
        
        encoded_cols = []
        for col in categorical_cols:
            if col in self.df.columns:
                le = LabelEncoder()
                self.df[f'{col}_encoded'] = le.fit_transform(self.df[col].astype(str))
                self.encoders[col] = le
                encoded_cols.append(f'{col}_encoded')
        
        self.preprocessed = True
        print("\nData preprocessing completed!")
        print(f"Dataset shape: {self.df.shape}")
        print(f"New features: Hour, DayOfWeek, Month, Violent_Crime, etc.")
        print(f"Encoded columns: {encoded_cols}")
        
        # Show sample of processed data
        print("\nSample of processed data:")
        sample_cols = []
        for col in ['City', 'Crime Description', 'Victim Age', 'Hour', 'Violent_Crime']:
            if col in self.df.columns:
                sample_cols.append(col)
        
        if sample_cols:
            display(self.df[sample_cols].head())
        else:
            print("No sample columns available")
    
    def exploratory_analysis(self):
        """Perform exploratory data analysis"""
        if not self.preprocessed:
            self.preprocess_data()
            
        # Check if data is available
        if self.df.empty:
            print("No data available for analysis")
            return
            
        print("\n=== EXPLORATORY DATA ANALYSIS ===\n")
        
        # Basic statistics
        print("Dataset Shape:", self.df.shape)
        print("\nMissing Values:")
        print(self.df.isnull().sum())
        
        # Create subplots
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Crime distribution by type
        if 'Crime Domain' in self.df.columns:
            self.df['Crime Domain'].value_counts().plot(kind='bar', ax=axes[0,0])
            axes[0,0].set_title('Crime Distribution by Domain')
            axes[0,0].tick_params(axis='x', rotation=45)
        
        # Top cities
        if 'City' in self.df.columns:
            self.df['City'].value_counts().head(10).plot(kind='bar', ax=axes[0,1])
            axes[0,1].set_title('Top 10 Cities by Crime Count')
            axes[0,1].tick_params(axis='x', rotation=45)
        
        # Crime by hour
        if 'Hour' in self.df.columns:
            self.df['Hour'].hist(bins=24, ax=axes[1,0])
            axes[1,0].set_title('Crime Distribution by Hour')
            axes[1,0].set_xlabel('Hour of Day')
        
        # Victim age distribution
        if 'Victim Age' in self.df.columns:
            self.df['Victim Age'].hist(bins=20, ax=axes[1,1])
            axes[1,1].set_title('Victim Age Distribution')
            axes[1,1].set_xlabel('Age')
        
        plt.tight_layout()
        plt.show()
        
        # Temporal analysis
        print("\n=== TEMPORAL ANALYSIS ===")
        if 'Month' in self.df.columns:
            monthly_crimes = self.df.groupby('Month').size()
            print("Monthly Crime Distribution:")
            print(monthly_crimes)
        
        if 'DayOfWeek' in self.df.columns:
            daily_crimes = self.df.groupby('DayOfWeek').size()
            days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
            print("\nDaily Crime Distribution:")
            for i, count in enumerate(daily_crimes):
                if i < len(days):
                    print(f"{days[i]}: {count}")
    
    def identify_crime_hotspots(self):
        """Identify crime hotspots using clustering"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for hotspot analysis")
            return None
            
        print("\n=== CRIME HOTSPOT IDENTIFICATION ===\n")
        
        # Group by city for hotspot analysis
        if 'City' not in self.df.columns:
            print("'City' column not found for hotspot analysis")
            return None
            
        city_crime_stats = self.df.groupby('City').agg({
            'Report Number': 'count',
            'Violent_Crime': 'mean',
            'Victim Age': 'mean',
            'Police Deployed': 'mean'
        }).rename(columns={
            'Report Number': 'Total_Crimes',
            'Violent_Crime': 'Violent_Crime_Ratio',
            'Victim Age': 'Avg_Victim_Age',
            'Police Deployed': 'Avg_Police_Deployed'
        }).sort_values('Total_Crimes', ascending=False)
        
        print("Top Crime Hotspots (Cities):")
        display(city_crime_stats.head(10))
        
        # Visualize top crime cities
        plt.figure(figsize=(12, 6))
        city_crime_stats.head(10)['Total_Crimes'].plot(kind='bar')
        plt.title('Top 10 Crime Hotspot Cities')
        plt.xlabel('City')
        plt.ylabel('Number of Crimes')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
        
        return city_crime_stats
    
    def crime_pattern_analysis(self):
        """Analyze crime patterns and correlations"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for pattern analysis")
            return None
            
        print("\n=== CRIME PATTERN ANALYSIS ===\n")
        
        # Correlation analysis
        numeric_cols = ['Victim Age', 'Police Deployed', 'Hour', 'DayOfWeek', 'Month', 'Violent_Crime']
        available_numeric_cols = [col for col in numeric_cols if col in self.df.columns]
        
        if len(available_numeric_cols) < 2:
            print("Not enough numeric columns for correlation analysis")
            return None
            
        correlation_matrix = self.df[available_numeric_cols].corr()
        
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
        plt.title('Crime Data Correlation Matrix')
        plt.tight_layout()
        plt.show()
        
        # Crime patterns by time and type
        print("\nCrime Patterns by Hour and Type:")
        if 'Hour' in self.df.columns and 'Violent_Crime' in self.df.columns:
            hour_crime_type = pd.crosstab(self.df['Hour'], self.df['Violent_Crime'], normalize='index')
            display(hour_crime_type.head())
        
        return correlation_matrix
    
    def predict_crime_occurrence(self):
        """Build predictive model for crime occurrence with proper NaN handling"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for prediction")
            return None, None
            
        print("\n=== CRIME PREDICTION MODEL ===\n")
        
        # Prepare features for prediction
        feature_columns = [
            'Victim Age', 'Police Deployed', 'Hour', 'DayOfWeek', 'Month',
            'City_encoded', 'Crime Description_encoded', 'Victim Gender_encoded',
            'Weapon Used_encoded'
        ]
        
        # Check which features are available
        available_features = [col for col in feature_columns if col in self.df.columns]
        
        if len(available_features) < 3:
            print("Not enough features for prediction model")
            return None, None
        
        # Check for missing values before proceeding
        print("Checking for missing values in features:")
        missing_summary = self.df[available_features].isnull().sum()
        print(missing_summary[missing_summary > 0])
        
        # Handle missing values - drop rows with NaN in features or target
        X = self.df[available_features]
        y = self.df['Violent_Crime']
        
        # Combine features and target to identify complete cases
        combined = pd.concat([X, y], axis=1)
        complete_cases = combined.dropna()
        
        print(f"Original dataset size: {len(X)}")
        print(f"Complete cases (no missing values): {len(complete_cases)}")
        print(f"Rows dropped due to missing values: {len(X) - len(complete_cases)}")
        
        # Update X and y with complete cases only
        X_clean = complete_cases[available_features]
        y_clean = complete_cases['Violent_Crime']
        
        # Check if we have enough data
        if len(X_clean) < 100:
            print("Not enough complete data for model training. Using imputation instead.")
            # Alternative: Use imputation
            imputer = SimpleImputer(strategy='median')
            X_clean = imputer.fit_transform(X)
            y_clean = y
        else:
            print("Sufficient complete data for model training")
        
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X_clean, y_clean, test_size=0.3, random_state=42, stratify=y_clean
        )
        
        print(f"Training set size: {len(X_train)}")
        print(f"Test set size: {len(X_test)}")
        
        # Train Random Forest classifier
        rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
        rf_model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = rf_model.predict(X_test)
        
        # Evaluate model
        print("\nModel Performance:")
        print("Classification Report:")
        print(classification_report(y_test, y_pred))
        
        # Feature importance
        feature_importance = pd.DataFrame({
            'feature': available_features,
            'importance': rf_model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        print("\nTop 10 Most Important Features for Crime Prediction:")
        display(feature_importance.head(10))
        
        # Plot feature importance
        plt.figure(figsize=(10, 6))
        sns.barplot(data=feature_importance.head(10), x='importance', y='feature')
        plt.title('Top 10 Feature Importance for Crime Prediction')
        plt.tight_layout()
        plt.show()
        
        # Confusion Matrix
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                    xticklabels=['Non-Violent', 'Violent'], 
                    yticklabels=['Non-Violent', 'Violent'])
        plt.title('Confusion Matrix - Violent Crime Prediction')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return rf_model, feature_importance

    def group_crimes_by_location(self):
        """Group crimes by geographical location and characteristics"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for location grouping")
            return None
            
        print("\n=== CRIME GROUPING BY LOCATION ===\n")
        
        # Create location-based crime groups
        if 'City' not in self.df.columns:
            print("'City' column not found for location grouping")
            return None
            
        location_groups = self.df.groupby('City').agg({
            'Report Number': 'count',
            'Violent_Crime': 'mean',
            'Crime Description': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
            'Weapon Used': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
            'Police Deployed': 'mean'
        }).rename(columns={
            'Report Number': 'Crime_Count',
            'Violent_Crime': 'Violent_Crime_Ratio',
            'Crime Description': 'Most_Common_Crime',
            'Weapon Used': 'Most_Common_Weapon',
            'Police Deployed': 'Avg_Police_Deployed'
        }).sort_values('Crime_Count', ascending=False)
        
        print("Crime Groups by Location (Top 15):")
        display(location_groups.head(15))
        
        return location_groups
    
    def generate_policy_recommendations(self):
        """Generate data-driven policy recommendations"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for policy recommendations")
            return []
            
        print("\n=== POLICY RECOMMENDATIONS ===\n")
        
        recommendations = []
        
        # 1. Resource allocation based on hotspots
        hotspots = self.identify_crime_hotspots()
        if hotspots is not None and not hotspots.empty:
            high_crime_cities = hotspots.head(5).index.tolist()
            recommendations.append(
                f"RESOURCE ALLOCATION: Focus police resources on high-crime cities: {', '.join(high_crime_cities)}"
            )
        
        # 2. Temporal patterns
        if 'Hour' in self.df.columns:
            peak_hours = self.df['Hour'].value_counts().head(3).index.tolist()
            recommendations.append(
                f"TEMPORAL DEPLOYMENT: Increase patrols during peak crime hours: {peak_hours}"
            )
        
        # 3. Crime-type specific strategies
        if 'Weapon Used' in self.df.columns and 'Violent_Crime' in self.df.columns:
            violent_weapons = self.df[self.df['Violent_Crime'] == 1]['Weapon Used'].value_counts().head(3).index.tolist()
            recommendations.append(
                f"WEAPON CONTROL: Focus on controlling: {', '.join(violent_weapons)}"
            )
        
        # 4. Victim demographics
        if 'Victim Age' in self.df.columns and 'Victim Gender' in self.df.columns:
            avg_victim_age = self.df['Victim Age'].mean()
            common_victim_gender = self.df['Victim Gender'].value_counts().index[0]
            recommendations.append(
                f"VICTIM PROTECTION: Target protection for {common_victim_gender} victims (avg age: {avg_victim_age:.1f})"
            )
        
        # 5. Preventive measures based on crime patterns
        if 'DayOfWeek' in self.df.columns and self.df['DayOfWeek'].notna().any():
            common_day_mode = self.df['DayOfWeek'].mode()
            if len(common_day_mode) > 0:
                common_day = int(common_day_mode[0])
                days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
                if 0 <= common_day < len(days):
                    recommendations.append(
                        f"PREVENTIVE PATROLLING: Increase community policing on {days[common_day]}s"
                    )
        
        # 6. Additional recommendations based on data insights
        if 'Case Closed' in self.df.columns:
            clearance_rate = self.df['Case Closed'].value_counts(normalize=True).get('Yes', 0)
            if clearance_rate < 0.3:
                recommendations.append(
                    f"INVESTIGATION IMPROVEMENT: Enhance investigative capabilities (current clearance rate: {clearance_rate:.1%})"
                )
        
        # Violent crime percentage
        if 'Violent_Crime' in self.df.columns:
            violent_percentage = self.df['Violent_Crime'].mean()
            if violent_percentage > 0.3:
                recommendations.append(
                    f"VIOLENT CRIME FOCUS: Implement special task force for violent crimes ({violent_percentage:.1%} of total)"
                )
        
        print("DATA-DRIVEN POLICY RECOMMENDATIONS:")
        for i, rec in enumerate(recommendations, 1):
            print(f"{i}. {rec}")
        
        return recommendations
    
    def calculate_precision_matrix(self):
        """Calculate and analyze the precision matrix for crime data"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for precision matrix analysis")
            return None, None
            
        print("\n=== PRECISION MATRIX ANALYSIS ===\n")
        
        # Select numeric features for precision matrix calculation
        numeric_features = ['Victim Age', 'Police Deployed', 'Hour', 'DayOfWeek', 'Month', 'Violent_Crime']
        
        # Filter out features with too many missing values
        available_features = []
        for feature in numeric_features:
            if feature in self.df.columns and self.df[feature].notna().sum() > 1000:  # At least 1000 non-null values
                available_features.append(feature)
        
        print(f"Features used for precision matrix: {available_features}")
        
        if len(available_features) < 2:
            print("Not enough features for precision matrix calculation")
            return None, None
        
        # Drop rows with missing values in selected features
        data_for_precision = self.df[available_features].dropna()
        
        print(f"Samples used for precision matrix: {len(data_for_precision)}")
        
        if len(data_for_precision) < 100:
            print("Not enough data for precision matrix calculation")
            return None, None
        
        # Standardize the data (important for precision matrix)
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data_for_precision)
        
        # Calculate covariance matrix
        covariance_matrix = np.cov(data_scaled, rowvar=False)
        print(f"Covariance matrix shape: {covariance_matrix.shape}")
        
        try:
            # Calculate precision matrix (inverse of covariance matrix)
            precision_matrix = np.linalg.inv(covariance_matrix)
            print("Precision matrix calculated successfully")
            
            # Create a DataFrame for better visualization
            precision_df = pd.DataFrame(
                precision_matrix,
                index=available_features,
                columns=available_features
            )
            
            # Display the precision matrix
            print("\nPrecision Matrix (Inverse of Covariance Matrix):")
            display(precision_df)
            
            # Visualize the precision matrix
            plt.figure(figsize=(10, 8))
            sns.heatmap(precision_df, annot=True, cmap='RdBu_r', center=0, fmt='.3f',
                       square=True, cbar_kws={'shrink': 0.8})
            plt.title('Precision Matrix (Inverse Covariance Matrix)')
            plt.tight_layout()
            plt.show()
            
            # Interpret the precision matrix
            print("\nINTERPRETATION OF PRECISION MATRIX:")
            print(" Diagonal elements: Precision of each variable (inverse of variance)")
            print(" Off-diagonal elements: Conditional relationships between variables")
            print(" Positive values: Variables increase together when controlling for others")
            print(" Negative values: Variables move in opposite directions when controlling for others")
            print(" Zero/Near-zero: Variables are conditionally independent")
            
            # Analyze strongest conditional relationships
            print("\nSTRONGEST CONDITIONAL RELATIONSHIPS:")
            relationships = []
            
            for i in range(len(available_features)):
                for j in range(i+1, len(available_features)):
                    if abs(precision_matrix[i, j]) > 0.1:  # Threshold for "strong" relationship
                        relationships.append({
                            'Variable 1': available_features[i],
                            'Variable 2': available_features[j],
                            'Precision Value': precision_matrix[i, j],
                            'Strength': abs(precision_matrix[i, j])
                        })
            
            if relationships:
                relationships_df = pd.DataFrame(relationships).sort_values('Strength', ascending=False)
                display(relationships_df.head(10))
            else:
                print("No strong conditional relationships found (all < 0.1)")
            
            return precision_matrix, precision_df
            
        except np.linalg.LinAlgError as e:
            print(f"Error calculating precision matrix: {e}")
            print("This usually happens when the covariance matrix is singular (not invertible)")
            print("Possible reasons: Highly correlated variables or insufficient data")
            
            # Alternative: Use pseudo-inverse or regularized precision matrix
            print("\nTrying regularized precision matrix...")
            try:
                # Use L2 regularization to make matrix invertible
                ridge_covariance = covariance_matrix + 0.1 * np.eye(covariance_matrix.shape[0])
                precision_matrix_regularized = np.linalg.inv(ridge_covariance)
                
                precision_df_regularized = pd.DataFrame(
                    precision_matrix_regularized,
                    index=available_features,
                    columns=available_features
                )
                
                print("Regularized precision matrix calculated")
                display(precision_df_regularized)
                
                return precision_matrix_regularized, precision_df_regularized
                
            except Exception as e2:
                print(f"Even regularized approach failed: {e2}")
                return None, None
    
    def advanced_precision_analysis(self):
        """Advanced precision matrix analysis with graphical models"""
        if not self.preprocessed:
            self.preprocess_data()
            
        if self.df.empty:
            print("No data available for advanced precision analysis")
            return None
            
        print("\n=== ADVANCED PRECISION MATRIX ANALYSIS ===\n")
        
        # Use graphical lasso for sparse precision matrix estimation
        try:
            numeric_features = ['Victim Age', 'Police Deployed', 'Hour', 'Violent_Crime']
            available_features = [f for f in numeric_features if f in self.df.columns]
            
            data_for_glasso = self.df[available_features].dropna()
            
            if len(data_for_glasso) < 100:
                print("Not enough data for graphical lasso")
                return None
            
            # Standardize data
            scaler = StandardScaler()
            data_scaled = scaler.fit_transform(data_for_glasso)
            
            # Apply Graphical Lasso
            gl = GraphicalLasso(alpha=0.1)  # Regularization parameter
            gl.fit(data_scaled)
            
            precision_glasso = gl.precision_
            
            precision_glasso_df = pd.DataFrame(
                precision_glasso,
                index=available_features,
                columns=available_features
            )
            
            print("Graphical Lasso Precision Matrix (Sparse):")
            display(precision_glasso_df)
            
            # Visualize the sparse precision matrix
            plt.figure(figsize=(8, 6))
            sns.heatmap(precision_glasso_df, annot=True, cmap='RdBu_r', center=0, fmt='.3f')
            plt.title('Graphical Lasso Precision Matrix (Sparse Conditional Dependencies)')
            plt.tight_layout()
            plt.show()
            
            print("\nGRAPHICAL LASSO INTERPRETATION:")
            print(" Non-zero entries indicate direct conditional dependencies")
            print(" Zero entries indicate conditional independence")
            print(" Sparse matrix reveals the underlying graph structure")
            
            return precision_glasso_df
            
        except ImportError:
            print("scikit-learn version doesn't support GraphicalLasso")
            print("Try: pip install scikit-learn")
            return None

print("CrimeAnalysis class definition completed with all methods")

# ===== MAIN EXECUTION =====
if __name__ == "__main__":
    # Initialize and run the complete analysis
    crime_analyzer = CrimeAnalysis('crime_dataset_india.csv')
    
    # Only run analysis if data was loaded successfully
    if not crime_analyzer.df.empty:
        # Run complete analysis
        print("\nStarting Comprehensive Crime Analysis...")
        print("="*60)

        # Run all analyses
        crime_analyzer.preprocess_data()
        crime_analyzer.exploratory_analysis()
        crime_analyzer.crime_pattern_analysis()
        crime_analyzer.predict_crime_occurrence()
        crime_analyzer.group_crimes_by_location()
        crime_analyzer.generate_policy_recommendations()

        # Run precision matrix analysis
        print("\n" + "="*60)
        print("Running Precision Matrix Analysis...")
        print("="*60)
        crime_analyzer.calculate_precision_matrix()
        crime_analyzer.advanced_precision_analysis()

        print("\n" + "="*60)
        print("COMPREHENSIVE CRIME ANALYSIS COMPLETED SUCCESSFULLY")
        print("="*60)
    else:
        print("\nAnalysis cannot proceed due to data loading issues")
```
